```python
import numpy as np
import matplotlib.pyplot as plt
from notebook_config import *
from scipy.optimize import curve_fit
```

```python
#Lab 2: Heat Capacity
time = [0, 30, 60, 90, 120, 150, 180, 240, 300, 260, 420, 480, 540, 600, 660, 720, 780, 840, 900, 960, 1020]
data_temp = [23.754, 23.769, 23.766, 23.760, 23.754, 23.749, 23.749, 24.238, 24.578, 24.816, 24.931, 24.986, 25.017, 25.023, 25.026, 25.023, 25.017, 25.012, 25.005, 24.996, 24.990]


def f(x, m, t):
    return m*x + t

popt1, pcov1 = curve_fit(f, time[:7], data_temp[:7])
popt2, pcov2 = curve_fit(f, time[12:], data_temp[12:])

y1 = [f(i,popt1[0],popt1[1]) for i in time]
y2 = [f(i,popt2[0],popt2[1]) for i in time]

jump = 180

plt.plot(time, data_temp, marker=".", linestyle="", label="Measurement")
plt.plot(time, y1,label="Linear fit of first part")
plt.plot(time, y2,label="Linear fit of second part")
plt.text(620,24.1,"Ignition at t=180s",fontsize=15)
plt.vlines(jump, f(jump, popt1[0], popt1[1]), f(jump, popt2[0], popt2[1]), linestyle="--", color="red")
plt.xlabel("Time (s)")
plt.ylabel("Temperature (C)")
plt.legend()
plt.title("Determining the heat capactiy")
plt.savefig("fig1.png", dpi=300)
plt.show()

diff = f(jump, popt2[0], popt2[1])- f(jump, popt1[0], popt1[1])

print("Red line (K): ~", diff)
```
