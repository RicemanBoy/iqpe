```python
from pyfiles.ftsteane import *
from HPC.Upload.classes import *
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Bitstream Vera Sans",
    "font.sans-serif": "Helvetica",
})
```

```python
shots=400

fig, ax1 = plt.subplots(figsize=(13,8))
ax1.xaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals = 2))

p, y, err = np.loadtxt("Simulationmatrix/nFT_RotSurf.txt", delimiter = ",")
err = [np.sqrt((i-i**2)/shots) for i in err]
ax1.plot(p,y, label="nFT RotSurf", marker = "",markersize = 20, alpha = 0.5)
ax1.errorbar(p, y, yerr=err, ecolor="red", linestyle="")

p1, y1, err1 = np.loadtxt("Simulationmatrix/Physicalqubits_5000shots.txt", delimiter = ",")
ax1.plot(p1,y1, label="Physical qubits", marker = "",markersize = 20, alpha = 0.5)
ax1.errorbar(p1, y1, yerr=err1, ecolor="red", linestyle="")

p2, y2, err2 = np.loadtxt("Simulationmatrix/nFT_RotSurf_qec.txt", delimiter = ",")
err2 = [np.sqrt((i-i**2)/shots) for i in err2]
ax1.plot(p2,y2, label="nFT RotSurf QEC", marker = "",markersize = 20, alpha = 0.5)
ax1.errorbar(p2, y2, yerr=err2, ecolor="red", linestyle="")

p3, y3, err3 = np.loadtxt("Simulationmatrix/nFT_RotSurf_PS.txt", delimiter = ",")
err3 = [np.sqrt((i-i**2)/shots) for i in err3]
ax1.plot(p3,y3, label="nFT RotSurf PS", marker = "",markersize = 20, alpha = 0.5)
ax1.errorbar(p3, y3, yerr=err3, ecolor="red", linestyle="")

p4, y4, err4 = np.loadtxt("Simulationmatrix/nFT_RotSurf_PS_qec.txt", delimiter = ",")
err4 = [np.sqrt((i-i**2)/shots) for i in err4]
ax1.plot(p4,y4, label="nFT RotSurf PS QEC", marker = "",markersize = 20, alpha = 0.5)
ax1.errorbar(p4, y4, yerr=err4, ecolor="red", linestyle="")

p5, y5, err5 = np.loadtxt("Simulationmatrix/nFT_steane.txt", delimiter = ",")
err5 = [np.sqrt((i-i**2)/shots) for i in err5]
ax1.plot(p5,y5, label="nFT Steane", marker="", linestyle="--", markersize = 20, alpha = 0.5)
ax1.errorbar(p5, y5, yerr=err5, ecolor="red", linestyle="")

p6, y6, err6 = np.loadtxt("Simulationmatrix/nFT_steane_qec.txt", delimiter = ",")
err6 = [np.sqrt((i-i**2)/shots) for i in err6]
ax1.plot(p6, y6, label="nFT Steane QEC", marker="", linestyle="--", markersize = 20, alpha = 0.5)
ax1.errorbar(p6, y6, yerr=err6, ecolor="red", linestyle="")

p7, y7, err7 = np.loadtxt("Simulationmatrix/nFT_steane_PS.txt", delimiter = ",")
err7 = [np.sqrt((i-i**2)/shots) for i in err7]
ax1.plot(p7, y7, label="nFT Steane PS", marker="", linestyle="--", markersize = 20, alpha = 0.5)
ax1.errorbar(p7, y7, yerr=err7, ecolor="red", linestyle="")

ax1.set_title(r"3 Iterations, 15 angles, each {} Shots".format(int(shots)), fontsize=20)
ax1.set_xlabel("Pauli Error probability", fontsize=25)
ax1.set_ylabel(r"Abs. difference to $\varphi$", fontsize=25)
ax1.tick_params(axis='both', which='major', labelsize=25)
ax1.grid()
#ax1.legend(fontsize=20)
fig.legend(loc="center left", bbox_to_anchor=(0.9, 0.5), fontsize=20)
#plt.tight_layout()
plt.show()
```
