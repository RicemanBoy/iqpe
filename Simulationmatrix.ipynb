```python
import textwrap


shots=400

fig, ax1 = plt.subplots(figsize=(10,8))
ax1.xaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals = 2))
#################### Physical Qubits ####################
p1, y1, err1 = np.loadtxt("Simulationmatrix/Physicalqubits_400shots.txt", delimiter = ",")
ax1.plot(p1,y1, label="Physical qubits", marker = "",markersize = 20, alpha = 0.5)
ax1.errorbar(p1, y1, yerr=err1, ecolor="red", linestyle="")
#################### nFT RotSurf ####################
p, y, err = np.loadtxt("Simulationmatrix/nFT_RotSurf.txt", delimiter = ",")
err = [i/np.sqrt(shots) for i in err]
ax1.plot(p,y, label="nFT RotSurf", marker = "",markersize = 20, alpha = 0.5)
ax1.errorbar(p, y, yerr=err, ecolor="red", linestyle="")

p2, y2, err2 = np.loadtxt("Simulationmatrix/nFT_RotSurf_qec.txt", delimiter = ",")
err2 = [i/np.sqrt(shots) for i in err2]
ax1.plot(p2,y2, label="nFT RotSurf QEC", marker = "",markersize = 20, alpha = 0.5)
ax1.errorbar(p2, y2, yerr=err2, ecolor="red", linestyle="")

p3, y3, err3 = np.loadtxt("Simulationmatrix/nFT_RotSurf_PS.txt", delimiter = ",")
err3 = [i/np.sqrt(shots) for i in err3]
ax1.plot(p3,y3, label="nFT RotSurf PS", marker = "",markersize = 20, alpha = 0.5)
ax1.errorbar(p3, y3, yerr=err3, ecolor="red", linestyle="")

p4, y4, err4 = np.loadtxt("Simulationmatrix/nFT_RotSurf_PS_qec.txt", delimiter = ",")
err4 = [i/np.sqrt(shots) for i in err4]
ax1.plot(p4,y4, label="nFT RotSurf PS QEC", marker = "",markersize = 20, alpha = 0.5)
ax1.errorbar(p4, y4, yerr=err4, ecolor="red", linestyle="")
#################### nFT Steane ####################
p5, y5, err5 = np.loadtxt("Simulationmatrix/nFT_steane.txt", delimiter = ",")
err5 = [i/np.sqrt(shots) for i in err5]
ax1.plot(p5,y5, label="nFT Steane", marker="", linestyle="--", markersize = 20, alpha = 0.5)
ax1.errorbar(p5, y5, yerr=err5, ecolor="red", linestyle="")

p6, y6, err6 = np.loadtxt("Simulationmatrix/nFT_steane_qec.txt", delimiter = ",")
err6 = [i/np.sqrt(shots) for i in err6]
ax1.plot(p6, y6, label="nFT Steane QEC", marker="", linestyle="--", markersize = 20, alpha = 0.5)
ax1.errorbar(p6, y6, yerr=err6, ecolor="red", linestyle="")

p7, y7, err7 = np.loadtxt("Simulationmatrix/nFT_steane_PS.txt", delimiter = ",")
err7 = [i/np.sqrt(shots) for i in err7]
ax1.plot(p7, y7, label="nFT Steane PS", marker="", linestyle="--", markersize = 20, alpha = 0.5)
ax1.errorbar(p7, y7, yerr=err7, ecolor="red", linestyle="")

p8, y8, err8 = np.loadtxt("Simulationmatrix/nFT_steane_PS_qec.txt", delimiter = ",")
err8 = [i/np.sqrt(shots) for i in err8]
ax1.plot(p8, y8, label="nFT Steane PS QEC", marker="", linestyle="--", markersize = 20, alpha = 0.5)
ax1.errorbar(p8, y8, yerr=err8, ecolor="red", linestyle="")
#################### FT Steane ####################
p9, y9, err9 = np.loadtxt("Simulationmatrix/FT_steane.txt", delimiter = ",")
err9 = [i/np.sqrt(shots) for i in err9]
ax1.plot(p9, y9, label="FT Steane", marker="", linestyle="-.", markersize = 20, alpha = 0.5)
ax1.errorbar(p9, y9, yerr=err9, ecolor="red", linestyle="")

p10, y10, err10 = np.loadtxt("Simulationmatrix/FT_steane_qec.txt", delimiter = ",")
err10 = [i/np.sqrt(shots) for i in err10]
ax1.plot(p10, y10, label="FT Steane QEC", marker="", linestyle="-.", markersize = 20, alpha = 0.5)
ax1.errorbar(p10, y10, yerr=err10, ecolor="red", linestyle="")

p11, y11, err11 = np.loadtxt("Simulationmatrix/FT_steane_PS.txt", delimiter = ",")
err11 = [i/np.sqrt(shots) for i in err11]
ax1.plot(p11, y11, label="FT Steane PS", marker="", linestyle="-.", markersize = 20, alpha = 0.5)
ax1.errorbar(p11, y11, yerr=err11, ecolor="red", linestyle="")

p12, y12, err12 = np.loadtxt("Simulationmatrix/FT_steane_PS_qec.txt", delimiter = ",")
err12 = [i/np.sqrt(shots) for i in err12]
ax1.plot(p12, y12, label="FT Steane PS QEC", marker="", linestyle="-.", markersize = 20, alpha = 0.5)
ax1.errorbar(p12, y12, yerr=err12, ecolor="red", linestyle="")

ax1.set_title(r"iQPE: 3 Iterations, 15 angles, each {} Shots".format(int(shots)), fontsize=20)
ax1.set_xlabel("Pauli Error probability", fontsize=25)
ax1.set_ylabel(r"Abs. difference to $\varphi$", fontsize=25)
ax1.tick_params(axis='both', which='major', labelsize=25)
ax1.grid()
#ax1.legend(fontsize=20)
fig.legend(loc="center left", bbox_to_anchor=(0.9, 0.5), fontsize=20)
# caption="In the case of QEC, we have one QEC cycle right before the readout and one after every 8th T/Tdg-gate for all the codes. In case of no Postselection (PS), we do a random coinflip if the readout does not yield a logical 0 or 1. This is also the case for FT Steane and its flags. If PS is turned on, we discard the shot and re-do it until we get a logical 0 or 1."
# wrapped_caption = textwrap.fill(caption, width=180)

# # Add caption
# fig.text(
#     0.65,
#     -0.09,
#     wrapped_caption,
#     ha='center',
#     va='bottom',
#     fontsize=12
# )

# plt.tight_layout()
plt.savefig("Simulationmatrix/plot.png", bbox_inches="tight", dpi=200)
plt.show()
```
