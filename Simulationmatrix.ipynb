```python
from pyfiles.ftsteane import *
from HPC.Upload.classes import *
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Bitstream Vera Sans",
    "font.sans-serif": "Helvetica",
})
```

```python
# Input files
file1 = "HPC/text/FT_steane_0_2.txt"
file2 = "HPC/text/FT_steane_3_5.txt"
file3 = "HPC/text/FT_steane_6_9.txt"

# Output file
output_file = "Simulationmatrix/FT_steane.txt"

with open(file1, "r") as f1, open(file2, "r") as f2, open(file3, "r") as f3, open(output_file, "w") as out:

    # Read both files line by line together
    for line1, line2, line3 in zip(f1, f2, f3):

        # Remove newline characters
        line1 = line1.strip()
        line2 = line2.strip()
        line3 = line3.strip()

        # Combine with comma separation
        combined_line = f"{line1},{line2},{line3}"

        # Write to output file
        out.write(combined_line + "\n")
```
